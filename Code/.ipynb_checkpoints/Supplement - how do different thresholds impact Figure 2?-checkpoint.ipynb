```python
total_turns_strangers = 0
flist = glob.glob(os.path.join(base_dir, 'Analyses', 'turn_taking', 'strangers_pure', '*.csv'))
for file in flist:
    data = pd.read_csv(file)
    total_turns_strangers += len(data)
    
total_turns_friends = 0
flist = glob.glob(os.path.join(base_dir, 'Analyses', 'turn_taking', 'friends', '*.csv'))
for file in flist:
    data = pd.read_csv(file)
    total_turns_friends += len(data)
    
```

```python
flist = glob.glob(os.path.join(base_dir, 'Analyses', 'different_long_gap_cutoffs', '*.csv'))

for file in flist:
    
    long_gaps_all = pd.read_csv(file)
    cutoff = file.split('_')[-1].split('.csv')[0]

    strangers = long_gaps_all.loc[long_gaps_all['condition'] == 'strangers'].reset_index(drop=True)
    friends = long_gaps_all.loc[long_gaps_all['condition'] == 'friends'].reset_index(drop=True)
    
    print('threshold = {}'.format(cutoff))
    print('strangers n = {}'.format(len(strangers) / 14))
    print('strangers percent = {}'.format((len(strangers) / 14) / total_turns_strangers))
    print('friends n = {}'.format(len(friends) / 14))
    print('friends percent = {}'.format((len(friends) / 14) / total_turns_friends))
    print('____')
    
    friend_color = '#F46036'
    stranger_color = '#0FA3B1'
    
    #BOTH
    plt.figure(figsize=(8,5))

    timepoint_labels = ['6 seconds\nbefore\n(baseline)', '4 seconds\nbefore', '2 seconds\nbefore', 'long gap',
                       '2 seconds\nafter', '4 seconds\nafter', '6 seconds\nafter']
    
    ax = sns.pointplot(x="timepoint", y="connection_change", hue="condition",
                   data=long_gaps_all, dodge=False,
                  order=["3_before", "2_before", "1_before", 'long_gap', '1_after', '2_after', '3_after'],
                  palette = [stranger_color,friend_color], markers='o', ci=None)

    plt.xlabel("timepoint", fontsize=18, labelpad=20)
    plt.ylabel("connection rating", fontsize=18, labelpad=10)
    plt.title("long gaps > {}ms".format(cutoff), fontsize=18)

    ax.tick_params(labelsize=13)
    ax.set(ylim=(-3, 1.5)) #SET COMMON SCALE AFTER INSPECTING BOUNDS

    ax.set_xticklabels(timepoint_labels)

    ax.axvspan(2.75, 3.25, alpha=0.3, color='silver')

    sns.despine()
    plt.tight_layout()

    plt.savefig(os.path.join(base_dir, 'Figures', 'Figure2B_Supplement', '{}.pdf'.format(cutoff)), dpi=300) 
    plt.close()
    
```
